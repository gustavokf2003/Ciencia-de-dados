```python
# ID do arquivo no Google Drive
file_id = "1atr5eJYq3fcgubt9X5pAoPZqzyjPsVBg"

# URL correta para download
url = f"https://drive.google.com/uc?id={file_id}"

# Nome do arquivo local para salvar
output = "clima.csv"

# Baixando o arquivo
gdown.download(url, output, quiet=False)

# Carregando no Pandas
df_tempo = pd.read_csv(output)

# Verificando as primeiras linhas
df_tempo
```

```python
df_tempo['date'] = pd.to_datetime(df_tempo['date'])
df_tempo['date'] = df_tempo['date'].dt.tz_localize(None)
df_tempo['km'] = df_tempo['km'].astype(float)
df_tempo
```

```python
def merge_acidente(df1, df2):
    df1 = df1.sort_values(by='km')
    df2 = df2.sort_values(by='km')
    df1 = pd.merge_asof(df1, df2, 
                           on='km', left_by=['datetime'], right_by=['date'],
                           direction='nearest', tolerance=8)  
 
    return df1
```

```python
df_train = merge_acidente(df_train, df_tempo)
df_test = merge_acidente(df_test, df_tempo)
df_train
```

```python
df_train.to_csv("df_train.csv", index=False)
df_test.to_csv("df_test.csv", index=False)
```
